# Testing two blood pressure drugs 

Testing if a new drug lowers blood pressure more effectively than an old drug.

**Hypothesis**
- Null Hypothesis (H₀): There is no difference in the blood pressure reduction between the new and old drugs.
- Alternative Hypothesis (H₁): The new drug lowers blood pressure more effectively than the old drug.

**Group Formation**
- We have randomly picked up patients from the each group being treated by old drug and new drug. 

**Sample Size**
- Sample size is 30. We checked the reduction level in 30 patients in each group.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Data for the two groups showing reduction in their blood pressure
# Sample size 30
old_drug = np.array([5, 7, 6, 4, 5, 6, 7, 5, 4, 6, 8, 2, 6, 8, 5, 5, 7, 6, 4, 5, 6, 7, 5, 4, 6, 8, 2, 6, 8, 5])
new_drug = np.array([8, 8, 9, 5, 8, 10, 7, 8, 6, 12, 8, 7, 6, 8, 7, 7, 9, 8, 6, 8, 9, 9, 8, 6, 10, 8, 8, 9, 8, 9])


# Perform a t-test for independent samples
t_stat, p_value = stats.ttest_ind(old_drug, new_drug)

# Print the results
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_value}")
print("The P value is much lower than : 0.05 .. hence the control is considered to be Significant")


# Means of the two groups
print()
mean_old_drug = np.mean(old_drug)
print("Mean - Old Drug = ", mean_old_drug)
mean_new_drug = np.mean(new_drug)
print("Mean - New Drug = ", mean_new_drug)

# Plotting
fig, ax = plt.subplots()

# Box plots
ax.boxplot([old_drug, new_drug], positions=[1, 2], patch_artist=True, 
           boxprops=dict(facecolor="lightgreen"), medianprops=dict(color="red"))

# Means as points
ax.plot(1, mean_old_drug, 'ro', label='Mean (Old Drug)')
ax.plot(2, mean_new_drug, 'ro', label='Mean (New Drug)')

# Adding titles and labels
ax.set_xticks([1, 2])
ax.set_xticklabels(['Old Drug', 'New Drug'])
ax.set_ylabel('Blood Pressure Reduction (mmHg)')
ax.set_title('Comparison of Blood Pressure Reduction with Old and New Drugs')

# Adding a legend
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
print("The plot is also showing that the box for the old drug is far lower than the box for the new one.")
```

### Conclusion
Now we have strong evidence to conclude that the new drug significantly reduces blood pressure more effectively than the old drug.
